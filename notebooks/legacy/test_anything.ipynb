```python
import pandas as pd

# Mixed formats + an invalid value
s = pd.Series([
    "2026-04-23",          # ISO date
    "04/23/2026",          # US style
    "23-04-2026",          # EU style
    "2026/04/23 15:30:00", # with time
    "Apr 23, 2026",        # text month
    "not-a-date",          # invalid
])

parsed = pd.to_datetime(s, errors="coerce", format="mixed")

print("Original:")
print(s.to_list())
print("\nParsed:")
print(parsed.to_list())
print("\nNaT count (failed parses):", parsed.isna().sum())
```
